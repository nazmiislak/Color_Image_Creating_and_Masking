```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
width , height , channel = 300 , 300, 3

my_red_img=np.zeros((width , height , channel),dtype=np.uint8)

my_red_img[: , : , 0]=255

plt.imshow(my_red_img)
plt.axis('off')
plt.title("RED IMAGE")
```

```python
my_green_img=np.zeros((width , height , channel),dtype=np.uint8)
my_green_img[: , : , 1]=255

my_blue_img=np.zeros((width , height , channel),dtype=np.uint8)
my_blue_img[: , : , 2]=255

fig=plt.figure(figsize=(10,7))
rows=1
columns=3
fig.add_subplot(rows,columns,1)

plt.imshow(my_red_img ,cmap="gray")
plt.axis('off')
plt.title("Red IMAGE")

fig.add_subplot(rows,columns,2)
plt.imshow(my_green_img,cmap="gray")
plt.axis('off')
plt.title("Green IMAGE")

fig.add_subplot(rows,columns,3)
plt.imshow(my_blue_img,cmap="gray")
plt.axis('off')
plt.title("Blue IMAGE")
```

```python
my_img_white=np.ones((width , height , channel),dtype=np.uint8)*255
my_img_black=np.ones((width , height , channel),dtype=np.uint8)

my_img_gray=np.random.randint(128,129,(width , height , channel),dtype=np.uint8)

fig=plt.figure(figsize=(10,7),facecolor='Purple')
rows=2
columns=2
fig.add_subplot(rows,columns,1)

plt.imshow(my_img_white ,cmap="gray")
plt.axis('off')
plt.title("White IMAGE")

fig.add_subplot(rows,columns,2)
plt.imshow(my_img_black,cmap="gray")
plt.axis('off')
plt.title("Black IMAGE")

fig.add_subplot(rows,columns,3)
plt.imshow(my_img_gray,cmap="gray")
plt.axis('off')
plt.title("Gray IMAGE")

fig.add_subplot(rows,columns,4)
plt.imshow(my_img_gray,cmap="gray")
plt.axis('off')
plt.title("Gray IMAGE")
```

```python
new_width , new_height = 300,450

flag_germany = np.zeros((new_width , new_height,channel),dtype=np.uint8)

flag_germany[0:100,:] = [0,0,0]

flag_germany[100:200,:] = [255,0,0]

flag_germany[200:300,:] = [255,255,0]

plt.imshow(flag_germany)
plt.axis('off')
plt.title("Germany Flag")
```

```python
new_width , new_height = 300,450

flag_belgium = np.zeros((new_width , new_height,channel),dtype=np.uint8)

flag_belgium[:,0:150] = [0,0,0]

flag_belgium[:,150:300] = [255,255,0]

flag_belgium[: , 300:450] = [255,0,0]


mask_yellow=(flag_belgium[:,:,0]==255)&(flag_belgium[:,:,1]==255)&(flag_belgium[:,:,2]==0)

flag_belgium[mask_yellow] = [0,0,255]

plt.imshow(flag_belgium)
plt.axis('off')
plt.title("Belgium Flag")
```
